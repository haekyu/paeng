## Class
- class는 복잡한 자료구조를 디자인하기 위해 씀!
- class의 구성 요소
    - Atrribute / Field
        - class의 내부 데이터
        - class 내에서 `self.변수이름` 으로 사용
        - class 밖에서 `object변수.변수이름` 으로 사용
    - Function / Method
        - 함수들
        - 첫 번째 인풋으로 self를 반드시 포함
        - class 내에서 `self.함수이름` 으로 사용
        - class 밖에서 `object변수.함수이름` 으로 사용
    - Constructor / Initializer
        - `def __init__(self): ...`
        - attribute들을 초기화

- 예)

```python

class Cat:
    # Attributes
    name = ''
    brain = {}


    def __init__(self):
        self.name = 'cat-name'
        print('Cat is born')

    # Get 류 함수
    def get_name(self):
        return self.name

    # Set 류 함수
    def set_name(self, new_name):
        old_name = self.get_name()
        self.name = new_name
        print('Name is changed from {} to {}'.format(old_name, new_name))

    # 그 외
    def eat(self, food, taste):
        self.brain[food] = taste

    def eat_many(self, foods, tastes):

        for food, taste in zip(foods, tastes):
            self.eat(food, taste)


foods = ['apple', 'banana', 'fish', 'water', 'apple']
tastes = ['good', 'bad', 'not bad', 'great', 'great']

v = Cat()
v.set_name('V')
v.eat_many(foods, tastes)
print(v.brain)
print(v.get_name())
```
출력 결과
```
Cat is born
Name is changed from cat-name to V
{'apple': 'good'}
V
```

## Dictionary 복습
- Initialization
    - `d = {}`
    - `d = defaultdict(lambda: 디폴트값)`

- 검색
    - `d[key]`
    - `d.get(key, 검색디폴트값)`

- 수정
    - d[key] = ?
    - 예) `d[key] = d[key] + 1`

- 추가
    - `d[new_key] = new_val`

## Git
- Why Git?
    - 코드의 version 관리
    - 매 수정마다 screenshot(commit) 을 남김
    - 협업 용이
    - 여러 사람들의 수정이 동시에 반영이 가능
    - 다른 사람들이 어떻게 수정했는지 볼 수 있음
- 용어
    - Local
        - 내 컴퓨터
        - client 느낌. Git server와 통신(?) 하는 내 컴퓨터
    - Repository
        - server 느낌. Git server의 코드 저장소
- Clone
    - Git server의 repository 를 local로 복사해오는 행위
    - local과 repository를 연동시킴
    - `git clone 클론할_레포지토리_주소`
- Status check
    - `git status` 항상 생활화 하자
- Add
    - local 내의 새로운 내용물을 git 이 트래킹할 수 있게 해줌
    - `git add 추가할_파일_혹은_디렉토리_경로`
    - `git add .` 하면 현재 디렉토리 기준 전체가 (하위 경로 포함) add 됨
- Commit
    - Git 에서의 screenshot
    - `git commit -m "커밋메세지"`
- Push
    - Local 에서 Git repository로 정보를 전달하는 행위
    - `git push`
- Pull
    - Git repository의 정보를 Local로 가져오는 행위
    - `git pull`

## Linux command
- `clear`
    - 화면 정리.
- `ls`
    - list의 줄임말. 현재 위치의 file list를 보여줌.
    - 옵션
        - `ls -a`
            - list all. 숨김파일까지 보여줌
        - `ls -l`
            - 만들어진 날짜 / 파일 크기 등 기타 정보도 보여줌.
        - `ls -t`
            - 시간순 배열
        - `ls -G`
            - 예쁘게 컬러링 해서 리스트 나옴
        - `ls -alt` 등으로 여러 옵션을 한꺼번에 줄 수 있다.
- `pwd`
    - 현재 위치의 절대경로를 알려줌
- `cd`
    - change directory. 내가 원하는 디렉토리로 이동
    - `cd 내가_이동하고_싶은_디렉토리_경로`
    - ex) `cd ~/Desktop` 하면 바탕화면으로 이동
    - cf) ~ 은 홈 화면을 의미. 
- `mkdir`
    - make directory. 디렉토리 생성
    - `mkdir 원하는_디렉토리_경로`
- `cp`
    - copy. 파일 복사
    - `cp from_path to_path`
    - ex) `cp love.txt ../love2.txt` 하면 현재 디렉토리의 love.txt를 상위 폴더의 love2.txt로 복사
    - 옵션
        - `cp -r from_path to_path`
            - 디렉토리 복사
            - 디렉토리를 복사하고싶으면 반드시 -r 옵션을 줘야 함.
- `mv`
    - move. 파일/디렉토리 이동
    - `move from_path to_path`
- `rm`
    - remove
    - `rm 파일경로`
    - 옵션
        - `rm -r 디렉토리`
            - 디렉토리 제거
        - `rm -i 파일`
            - 파일 삭제하기 전에 정말 삭제할 거냐고 yes or no 물어봄.
            - `.bash_profile` 내에서 alias를 통해 rm 을 하면 항상 rm -i가 실행되게끔 만들어 놓으면 좋음.
